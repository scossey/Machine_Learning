```python
from ray import tune
from ray.tune.search.hyperopt import HyperOptSearch
from ray.tune.schedulers import ASHAScheduler
from ray.air import session
from sklearn.metrics import f1_score


# Training function
def train_xgb(config):
    global xgb  # Use the globally extracted PyCaret model

    # Convert max_features to float if it's 'sqrt'
    if config["max_features"] == "sqrt":
        num_features = X_tr.shape[1]
        colsample_bytree = (num_features ** 0.5) / num_features
    else:
        colsample_bytree = config["max_features"]

    # Update parameters for hyperparameter tuning
    xgb.set_params(
        n_estimators=config["n_estimators"],
        max_depth=config["max_depth"],
        colsample_bytree=colsample_bytree
    )

    # Evaluate the updated model
    preds = xgb.predict(X_te)
    f1 = f1_score(y_test, preds)
    session.report({"f1_score": f1})

```

```python
# Run tuning with Ray Tune
search_space = {
    "n_estimators": tune.choice([50, 100, 300, 500, 800]),
    "max_depth": tune.choice([15, 20, 25, None]),
    "max_features": tune.choice(["sqrt", 0.3, 0.5, 0.7])
}
```

```python
search_alg = HyperOptSearch(metric="f1_score", mode="max")
scheduler = ASHAScheduler(metric="f1_score", mode="max")

tuner = tune.Tuner(
    train_xgb,
    param_space=search_space,
    tune_config=tune.TuneConfig(
        search_alg=search_alg,
        scheduler=scheduler,
        num_samples=25
    )
)

tuner
```

```python
tuned_xgb = tuner.fit()
```
